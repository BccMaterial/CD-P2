# Dependências 

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

# Load do Dataset

```python
df = pd.read_csv("../data/PRSA_data_2010.1.1-2014.12.31.csv")
display(df)
```

# Transformação dos Dados

```python
df["date"] = pd.to_datetime({
    "year": df["year"],
    "month": df["month"],
    "day": df["day"],
    "hour": df["hour"]
})
df.dropna(inplace=True)
```

# Plot da Série Temporal(Data x NM2.5)

```python
plt.plot(df["date"], df["pm2.5"])
plt.xlabel("Data")
plt.ylabel("Qtd. de Partículas Microscópicas")
```

# Teste de Estacionalidade

```python
adf_test = adfuller(df['pm2.5'])
print('ADF Statistic: %f' % adf_test[0])
print(f"p-value: {adf_test[1]:.20f}")
```
